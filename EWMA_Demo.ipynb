```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
for i in range(10):
    x = 3 * np.arange(0, 5, 0.01)
    y = 5 * np.sin(x)
    y += np.random.normal(0, 3, len(x))
    v = y[0]
    beta = 0.1 * i
    beta_bias = 0.3
    fitted_y = np.zeros_like(y)
    for j in range(len(y)):
        v = beta * v + (1 - beta) * y[j]
        v_bias = v / (1 - beta_bias**(i + 1))
        fitted_y[j] = v_bias
        
    plt.figure(i + 1)
    plt.figure(figsize=(20, 8))
    plt.suptitle(f'Beta = {beta}')

    plt.subplot(1, 3, 1)
    sns.scatterplot(x=x, y=y)

    plt.subplot(1, 3, 2)
    plt.plot(x, y)

    plt.subplot(1, 3, 3)
    sns.scatterplot(x=x, y=y)
    plt.plot(x, fitted_y, color='red')
    plt.plot(x, 5 * np.sin(x), color='black')

    plt.show()
```

```python
# sampling rate
sr = 2000
# sampling interval
ts = 1.0/sr
t = np.arange(0,1,ts)

freq = 10.
x = 3*np.sin(2*np.pi*freq*t)

noise = np.random.normal(0, 100, 2000)
x+= noise

plt.figure(figsize = (8, 6))
plt.plot(t, x, 'r')
plt.ylabel('Amplitude')

plt.show()


from numpy.fft import fft, ifft

X = fft(x)
N = len(X)
n = np.arange(N)
T = N/sr
freq = n/T 

plt.figure(figsize = (12, 6))
plt.subplot(121)

plt.stem(freq, np.abs(X), 'b', \
         markerfmt=" ", basefmt="-b")
plt.xlabel('Freq (Hz)')
plt.ylabel('FFT Amplitude |X(freq)|')
plt.xlim(0, 15)

plt.subplot(122)
plt.plot(t, ifft(X), 'r')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.tight_layout()
plt.show()
```
